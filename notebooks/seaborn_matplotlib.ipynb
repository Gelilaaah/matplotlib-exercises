```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years = range(2000, 2008)
apples = [0.895, 0.919, 0.926, 0.929, 0.929, 0.931, 0.934, 0.936]
oranges = [0.921, 0.878, 0.943, 0.919, 0.904, 0.908, 0.923, 0.962]
```

```python
sns.set_style('darkgrid')
plt.plot(years, apples, marker = '*', c='darkmagenta')
plt.plot(years, oranges, marker = '.', c='teal', ls='dashed')
plt.title('Fruits Yield')
plt.xlabel('Years')
plt.ylabel('Yield')
plt.legend(['apples', 'oranges'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')

flowers_df = sns.load_dataset('iris')
sns.scatterplot(data = flowers_df, x = 'sepal_length', y = 'sepal_width', hue = 'species');
```

# Histogram
used to show the distribution of data

```python
import numpy as np
plt.title('Distribution of sepal width')
plt.hist(flowers_df.sepal_width, bins = np.arange(2, 5, 0.25));
```

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']

plt.hist(setosa_df.sepal_width, alpha= 0.4, bins = np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha= 0.4, bins = np.arange(2, 5, 0.25));
plt.legend(['setosa', 'versicolor'])
```

```python
import numpy as np
plt.title('Distribution of sepal width')
plt.hist([setosa_df.sepal_width, virginica_df.sepal_width, versicolor_df.sepal_width], bins = np.arange(2, 5, 0.25), stacked = True); ## they have to be passed as a list!!
plt.legend(['Setosa', 'Virginica', 'Versicolor'])
```

# Bar chart
Similar to line charts but a bar chart is shown for each value rather than points connected by lines

```python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
plt.bar(years, apples);
```

```python
plt.bar(years, apples)
plt.bar(years, oranges, bottom = apples);
plt.legend(['Apples', 'Oranges'])
```

```python
plt.bar(years, oranges);
plt.plot(years, oranges, 'o--r');
plt.title('Yield of oranges')
```

# Heatmap
used to visualize a 2d data like a matrix or a table using colors

```python
flights_df = sns.load_dataset('flights').pivot(index = 'month', columns = 'year', values = 'passengers')
flights_df ## use pivot() to convert the df into a matrix format
```

```python
sns.heatmap(flights_df);
```

```python
plt.title('No. of passengers (1000s)')
sns.heatmap(flights_df, fmt= 'd', annot = True, cmap='Blues');
```

# Plotting multiple charts in a grid

```python
fif, axes = plt.subplots(2,3, figsize=(16, 8))
axes[0, 0].plot(years, apples, 's-b')
axes[0, 0].plot(years, oranges, 'o--r')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Yield')
axes[0, 0].set_title('Crop Yields')
axes[0, 0].legend(['Apples', 'Oranges'])

axes[0,1].set_title('Sepal length vs width')
sns.scatterplot(data = flowers_df, x = 'sepal_length', y = 'sepal_width', hue = 'species', ax=axes[0,1]);

axes[0,2].hist([setosa_df.sepal_width, virginica_df.sepal_width, versicolor_df.sepal_width], bins = np.arange(2, 5, 0.25), stacked = True);
axes[0,2].set_title('Distribution of sepal width')

axes[1, 0].set_title('Restaurant bills')
sns.barplot(data=tips_df, x ='day', y = 'total_bill', hue = 'sex', ax=axes[1,0]);

axes[1,1].set_title('No. of Passengers')
sns.heatmap(flights_df, fmt='d', annot = True, cmap='Blues', ax=axes[1,1]);
```
