```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
yield_a = [0.895, 0.91, 0.919]
plt.plot(yield_a)
```

```python
plt.plot(yield_a, marker = '.')
```

```python
plt.xlabel('estimates')
plt.title('yield of apples')
plt.plot(yield_a)
```

```python
years = range(2000, 2008)
apples = [0.895, 0.919, 0.926, 0.929, 0.929, 0.931, 0.934, 0.936]
oranges = [0.921, 0.878, 0.943, 0.919, 0.904, 0.908, 0.923, 0.962]
```

```python
sns.set_style('darkgrid')
plt.plot(years, apples, marker = '*', c='darkmagenta')
plt.plot(years, oranges, marker = '.', c='teal', ls='dashed')
plt.title('Fruits Yield')
plt.xlabel('Years')
plt.ylabel('Yield')
plt.legend(['apples', 'oranges'])
```

# Scatter Plot
- is used to visualize the relationship between two variables as points on a 2d grid

```python
flowers_df = sns.load_dataset('iris')
flowers_df
```

```python
plt.plot(flowers_df['sepal_length'], flowers_df['sepal_width'] );
```

hard to plot using a line plot so we use a scatter plot instead

```python
sns.scatterplot(flowers_df.sepal_length);
```

```python
sns.scatterplot(flowers_df.sepal_width);
```

```python
sns.scatterplot(flowers_df.sepal_length, flowers_df.sepal_width);
```
